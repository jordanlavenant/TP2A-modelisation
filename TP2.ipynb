```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import get_cmap
```

## Création d'une image 2x2

```python
image = np.array([[0, 1],
                  [1, 0]])
```

## Affichage

```python
cmp = get_cmap('gray')
plt.imshow(image, cmap=cmp, vmin=0, vmax=1)
plt.show()
```

## Applatissement

```python
image_flattened = image.flatten()
print(image_flattened)
```

## Ajout de bruit

```python
noise = np.random.normal(0, 0.1, image.shape)
image_noisy = np.clip(image + noise, 0, 1)
```

## Affichage côte à côte

```python
plt.imshow(image_noisy, cmap=cmp, vmin=0, vmax=1)
plt.show()
```

## Exercice 1

```python
# Génération
image = np.array([[0.9, 0.2],
                  [0.1, 0.8]])
```

```python
# Affichage
cmp = get_cmap('gray')
plt.imshow(image, cmap=cmp, vmin=0, vmax=1)
plt.show()
```

```python
# Applatissement
image_flattened = image.flatten()
print(image_flattened)
```

```python
# Scores
z = [1.2, 0.5]
```

```python
def softmax(z):
    return np.exp(z) / np.sum(np.exp(z))
```

```python
# Classe prédite par le réseau
(q1, q2) = softmax(z)
print(q1, q2)
```

## Exercice 2

```python
# Entropie croisée catégorielle
def ecc(p, q):
    return -np.sum(p * np.log(q))
```

```python
# Mon interprétation
p = [1, 0]
# R.N
q = [q1, q2]

ecc(p, q)
```

> Il y a 0.403 d'erreur ce qui est peu, notre réseau est donc bien entrainé.
